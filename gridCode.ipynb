```python
import random
import math
#for close coordinates for the endpoint
def generate_close_coordinates(lat, lon, max_distance_meters=2):
    # Conversion factor: 1 degree of latitude ≈ 111,139 meters
    lat_offset = max_distance_meters / 111139

    # Conversion factor for longitude depends on latitude
    lon_offset = max_distance_meters / (111139 * abs(math.cos(math.radians(lat))))

    # Generate two random offsets within the range
    new_coords = []
    for _ in range(2):
        new_lat = lat + random.uniform(-lat_offset, lat_offset)
        new_lon = lon + random.uniform(-lon_offset, lon_offset)
        new_coords.append((new_lat, new_lon))

    return new_coords

# Example usage
latitude = 7.56022
longitude = 9.02422
close_coordinates = generate_close_coordinates(latitude, longitude)
print(close_coordinates)
```
